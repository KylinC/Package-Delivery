```python
import networkx as nx
```

```python
import numpy as np
import pandas as pd 
import csv
```

```python
tools = pd.read_csv("tools.csv")
tools = pd.DataFrame(tools)
tools
```

```python
G = nx.MultiDiGraph()
```

```python
tools.loc[0]
```

```python
G.clear() 

for row in tools.itertuples():
    dep_city = getattr(row, "_1")
    arr_city = getattr(row, "_2")
    dep_time = getattr(row, "_3")
    arr_time = getattr(row, "_4")
    cost = getattr(row, "_5")

    G.add_edge(dep_city, arr_city, dep_time=dep_time, arr_time=arr_time, cost=cost)
```

```python
G[1][4][0]
```

```python
G[1][4][1]
```

```python
T = nx.dfs_tree(G, source=1, depth_limit=3)
list(T)
```

```python
E = nx.dfs_edges(G, source=1, depth_limit=400)
list(E)
```

```python
P = nx.all_simple_paths(G, 1, 5, cutoff=2)
list(P)
```
