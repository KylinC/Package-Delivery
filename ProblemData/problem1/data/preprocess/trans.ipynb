# xlsx to csv

```python
import pandas as pd
import numpy as np
import csv
```

```python
def xlsx_to_csv_pd(source_file, aim_file, sheet_name=[], aim_file_pile=[]):
    if (not sheet_name):
        data_xls = pd.read_excel(source_file, index_col=0)
        data_xls.to_csv(aim_file)
    else:
        for i in range(len(sheet_name)):
            print(sheet_name[i])
            data_xls = pd.read_excel(
                source_file, index_col=0, sheetname=sheet_name[i])
            data_xls.to_csv(aim_file_pile[i])
```

```python
xlsx_list = [
    "TableA-Orders.xlsx", "TableB-Commodities.xlsx",
    "TableC-DistanceMatrix.xlsx",
    "TableD-TransportationTools.xlsx"
]
csv_list = [
    "TableA-Orders.csv", "TableB-Commodities.csv",
    "TableC-DistanceMatrix.csv",
    "TableD-TransportationTools.csv"
]
```

```python
import os

xlsx_path = "./raw_xlsx/"
csv_path = "./raw_csv/"

for i in range(3):
    if csv_list[i] not in os.listdir(csv_path):
        xlsx_to_csv_pd(xlsx_path + xlsx_list[i],csv_path + csv_list[i])
```

```python
vehicles = ["Plane", "Ship", "Train", "Truck"]
aim_file_pile = [
    "TableD-Plane.csv", "TableD-Ship.csv", "TableD-Train.csv",
    "TableD-Truck.csv"
]


if aim_file_pile[0] not in os.listdir(csv_path):
    xlsx_to_csv_pd(xlsx_path + xlsx_list[3], csv_path + csv_list[3], vehicles,
                   [csv_path + item for item in aim_file_pile])
```

# csv analysis

## str time to float time

```python
def Eval(Str):
    Lstrip = Str.lstrip('0')
    if Lstrip == "":
        return 0
    else:
        return eval(Lstrip)


def Time(Str):  # 09:45:12 -> float (min)
    if type(Str) != type(""):
        return Str
    Split = Str.split(':')
    return Eval(Split[0]) * 60 + Eval(Split[1]) + Eval(Split[2]) / 60


Time("00:45:12")
```

```python
orders_df['Time of Order'] = orders_df['Time of Order'].map(Time)
orders_df
```

```python
processed_csv_path = './processed_csv/'

if "orders.csv" not in os.listdir(processed_csv_path):
    orders_df.to_csv(processed_csv_path + "orders.csv", index=None)
```

```python
plane_csv = pd.read_csv(csv_path + "TableD-Plane.csv")
ship_csv = pd.read_csv(csv_path + "TableD-Ship.csv")
train_csv = pd.read_csv(csv_path + "TableD-Train.csv")
truck_csv = pd.read_csv(csv_path + "TableD-Truck.csv")
```

```python
plane_df = pd.DataFrame(plane_csv).sort_values(by=[
    "Departure city index", "Arrival city index", "Departure time",
    "Cost for 50km for every 1kg good "
])
ship_df = pd.DataFrame(ship_csv).sort_values(by=[
    "Departure city index", "Arrival city index", "Departure time",
    "Cost for 50km for every 1kg good "
])
train_df = pd.DataFrame(train_csv).sort_values(by=[
    "Departure city index", "Arrival city index", "Departure time",
    "Cost for 50km for every 1kg good "
])
truck_df = pd.DataFrame(truck_csv).sort_values(by=[
    "Departure city index", "Arrival city index", "Departure time",
    "Cost for 50km for every 1kg good "
])
```

```python
tools_df = [plane_df, train_df, truck_df, ship_df]
for tool_df in tools_df:
    tool_df["Departure time"] = tool_df["Departure time"].map(Time)
```

## arrival time & cost per unit weight

```python
dis = np.load("dis.npy")
dis
```

```python
def arrival_time(x):
    return x["Departure time"] + x["Average delay per trip (minutes)"] + 60 * dis[
        x["Departure city index"] - 1, x["Arrival city index"] - 1] / x["Speed (km/h)"]

for tool_df in tools_df:
    tool_df["Arrival time"] = arrival_time(tool_df)
```

```python
def unit_weight_cost(x):
    return dis[x["Departure city index"] - 1, x["Arrival city index"] - 1] / 50 * x["Cost for 50km for every 1kg good "]


for tool_df in tools_df:
    tool_df["Cost for every 1kg good"] = unit_weight_cost(tool_df)
```
