```python
plane_df["tool type"] = 0
train_df["tool type"] = 1 
truck_df["tool type"] = 2 
ship_df["tool type"] = 3 

tools_df = plane_df.append([train_df, truck_df, ship_df], ignore_index=True)
tools_df_cpy = tools_df.copy(deep=True)
tools_df_cpy["Departure time"] = tools_df_cpy["Departure time"].map(
    lambda x: x + 1440)
tools_df_cpy["Arrival time"] = tools_df_cpy["Arrival time"].map(
    lambda x: x + 1440)

tools_df = tools_df.append([tools_df_cpy], ignore_index=True)

tools_df = tools_df.sort_values(
    by=["Departure city index", "Arrival city index", "Departure time"])

del tools_df["Average delay per trip (minutes)"]
del tools_df["Speed (km/h)"]
del tools_df["Cost for 50km for every 1kg good "]

# tools_csv = pd.read_csv("./data/tools.csv")
# tools_df = pd.DataFrame(tools_csv)
# tools_df
```
