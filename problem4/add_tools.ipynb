```python
import pandas as pd 
import numpy as np 
import csv
```

```python
tools = pd.DataFrame(pd.read_csv("tools.csv"))
tools
```

```python
tools_cpy = tools.copy(deep=True)


tools_cpy["Departure time"] = tools["Departure time"].map(
    lambda x: x + 1440 * 2)
tools_cpy["Arrival time"] = tools["Arrival time"].map(
    lambda x: x + 1440 * 2)


tools = tools.append([tools_cpy], ignore_index=True)
tools = tools.sort_values(by=["Departure city index", "Arrival city index", "Departure time"])
```

```python
tools
```

```python
tools.to_csv("tools.csv", index=None)
```
